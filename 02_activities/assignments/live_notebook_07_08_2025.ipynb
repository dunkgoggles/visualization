```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75, 50)
x2 = np.array(["Mendel", "Kethoth", "Naoe", "Gundr", "Loafe"])
y2 = np.array([110, 180, 240, 99, 220])

fig, (ax1, ax2) = plt.subplots(ncols=2, nrows=1, figsize=(10, 4))

#ax1: Scatter plot 
ax1.scatter(x1, y1, color='teal')
ax1.set_title("Warrior Sword Scatter", fontsize=12)
ax1.set_xlabel("Index", fontsize=10)
ax1.set_ylabel("Value", fontsize=10)
ax1.grid(True, linestyle='--', alpha=0.5)
ax1.tick_params(axis='both', labelsize=8)

#ax2: Bar plot
ax2.bar(x2, y2, color='darkorange')
ax2.set_title("Magic Staff Power", fontsize=12)
ax2.set_ylabel("Power Level", fontsize=10)
ax2.tick_params(axis='x', labelsize=10, rotation=45)
ax2.tick_params(axis='y', labelsize=8)

#Hide top and right spines for a cleaner look
for ax in (ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

x = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

fig, ax = plt.subplots(figsize = (7,3))
ax.bar(x,y1, color = 'indigo')
ax.plot(x,y2, color = 'red')
```

```python
y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none")
```

```python
#Customizing errorbar appearance

y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor= "indigo",
            elinewidth= 4,
            capsize = 6,
            capthick= 4)
```

```python
#Errorevery

y2_sd = np.std(y2)

fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')

ax.errorbar(x, y2,
            yerr = y2_sd,
            fmt = "none",
            ecolor= "indigo",
            elinewidth= 4,
            capsize = 6,
            capthick= 4,
            errorevery= 2)
```

```python
#Using images from the internet

from PIL import Image #To open images
import requests #To get images from URLs
from io import BytesIO #To store images

response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(x, y2, color = 'red')

ax_image = fig.add_axes([0.1, # X coordinate
                         0.11, # Y coordinate
                         0.15, # Image width
                         0.35]) # Image height

ax_image.imshow(image)
ax_image.axis('off')

```
